3. "ì˜¤ë˜ë‹¬ë¦¬ê¸°"ì˜ ê°’ì„ Min-Max Scalingì„ í†µí•´ 0~1ì‚¬ì´ì˜ ê°’ìœ¼ë¡œ ë³€í™˜í•©ë‹ˆë‹¤.

```python
# ì˜¤ë˜ë‹¬ë¦¬ê¸° ê°’ì„ ìŠ¤ì¼€ì¼ë§í•˜ê³  ìŠ¤ì¼€ì¼ë§ ëœ ê°’ ì‚´í´ë³´ê¸°
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df["ì˜¤ë˜ë‹¬ë¦¬ê¸°"] = scaler.fit_transform(df[["ì˜¤ë˜ë‹¬ë¦¬ê¸°"]])
df["ì˜¤ë˜ë‹¬ë¦¬ê¸°"]
```

```python
# ë‹¤ë¥¸ ì¢…ëª© ìŠ¤ì¼€ì¼ë§

df["50M ë‹¬ë¦¬ê¸°"] = scaler.fit_transform(df[["50M ë‹¬ë¦¬ê¸°"]])
df["ì•‰ì•„ìœ—ëª¸ì•ìœ¼ë¡œêµ½íˆê¸°"] = scaler.fit_transform(df[["ì•‰ì•„ìœ—ëª¸ì•ìœ¼ë¡œêµ½íˆê¸°"]])
df["ìœ—ëª¸ë§ì•„ì˜¬ë¦¬ê¸°"] = scaler.fit_transform(df[["ìœ—ëª¸ë§ì•„ì˜¬ë¦¬ê¸°"]])
```

### **3. ë°ì´í„° ì¡°ì •í•˜ê¸°**

ë‹¤ë¥¸ ì¢…ëª©ê³¼ ë‹¬ë¦¬ ê°’ì´ í¬ë©´ í´ ìˆ˜ë¡ ì²´ë ¥ì´ ë‚®ë‹¤ëŠ” ê²ƒì„ ë³´ì—¬ì£¼ëŠ” ì¢…ëª©ì´ ìˆìŠµë‹ˆë‹¤.<br>

### **4. K-Means ì•Œê³ ë¦¬ì¦˜ìœ¼ë¡œ êµ°ì§‘í™”í•˜ê¸°**

K-Means ì•Œê³ ë¦¬ì¦˜ìœ¼ë¡œ êµ°ì§‘í™”í•˜ëŠ” ê³¼ì •ì€ ì•„ë˜ì™€ ê°™ìŠµë‹ˆë‹¤.<br>

-  sklearn ë¼ì´ë¸ŒëŸ¬ë¦¬ì˜ cluster ëª¨ë“ˆì—ì„œ KMeans í•¨ìˆ˜ë¥¼ ë¶ˆëŸ¬ì˜µë‹ˆë‹¤.
- êµ°ì§‘í™” ë°©ë²• ì¤‘ KMeansë¥¼ í™œìš©í•´ 2ê°œì˜ êµ°ì§‘ìœ¼ë¡œ ë¶„ë¥˜í•˜ê² ë‹¤ëŠ” ê²ƒì„ ì„ ì–¸í•˜ê³  kmeans ë³€ìˆ˜ì— ë„£ìŠµë‹ˆë‹¤.
- "ì˜¤ë˜ë‹¬ë¦¬ê¸°"ì˜ ê°’ì„ ë°”íƒ•ìœ¼ë¡œ 2ê°œì˜ êµ°ì§‘ìœ¼ë¡œ ë¶„ë¥˜í•˜ê³  cluster ë³€ìˆ˜ì— ë„£ìŠµë‹ˆë‹¤.
- ì—´ì„ ê¸°ì¤€(axis=1)ìœ¼ë¡œ í•˜ì—¬ í•™ìƒ ì´ë¦„ê³¼ êµ°ì§‘í™”ëœ ê°’ì„ í•©ì³ í•˜ë‚˜ì˜ í‘œë¡œ ë§Œë“­ë‹ˆë‹¤.

```python
# ì˜¤ë˜ë‹¬ë¦¬ê¸° ê¸°ë¡ì„ ê¸°ì¤€ìœ¼ë¡œ 2ê°œì˜ ê·¸ë£¹ìœ¼ë¡œ êµ°ì§‘í™”

from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=2) # KMean ê°ì²´ ë§Œë“¤ê¸°
cluster = kmeans.fit(df[["ì˜¤ë˜ë‹¬ë¦¬ê¸°"]]) # KMeans í•™ìŠµì‹œí‚¤ê¸°
pd.concat([df[["ì´ë¦„"]], pd.DataFrame(cluster.labels_)], axis=1) # KMeans ê²°ê³¼ ì›ë˜ ë°ì´í„°í”„ë ˆì„ì— í•©ì¹˜ê¸°
```

```python
# ëª¨ë“  ì¢…ëª©ì˜ ê¸°ë¡ì„ ê¸°ì¤€ìœ¼ë¡œ 6ê°œì˜ ê·¸ë£¹ìœ¼ë¡œ êµ°ì§‘í™”

from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=6)
cluster = kmeans.fit(df[["50M ë‹¬ë¦¬ê¸°", "ì•‰ì•„ìœ—ëª¸ì•ìœ¼ë¡œêµ½íˆê¸°", "ì˜¤ë˜ë‹¬ë¦¬ê¸°", "ìœ—ëª¸ë§ì•„ì˜¬ë¦¬ê¸°"]])
pd.concat([df["ì´ë¦„"], pd.DataFrame(cluster.labels_)], axis=1)
```
