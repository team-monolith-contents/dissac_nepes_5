ì‹¤ìŠµì—ì„œëŠ” ì§ì¥ì—ì„œì˜ ì—°ì°¨ì™€ ì—°ë´‰ì— ëŒ€í•œ ë°ì´í„°ì…‹ì„ ë¶„ì„í•˜ì—¬,  
ë‘ ë³€ìˆ˜ ê°„ ìƒê´€ê´€ê³„ë¥¼ ì„ í˜•íšŒê·€ë¡œ ë¶„ì„í•´ë³´ì•„ìš”.  

```python
# ë°ì´í„° ë¶ˆëŸ¬ì˜¤ê¸°
import pandas as pd

df = pd.read_csv("https://tmn-bucket-materials-all.s3.ap-northeast-2.amazonaws.com/data/6-9.+Salary_Data.csv")
df
```

```python
import numpy as np

# x(ì—°ì°¨), y(ì—°ë´‰) ë°ì´í„° ë¶„ë¦¬
x_data = np.array(df['YearsExperience'], dtype=np.float32).reshape((-1, 1)) 
y_data = np.array(df['Salary'], dtype=np.float32).reshape((-1, 1))

print(x_data.shape)
print(y_data.shape)
```

```python
# ë°ì´í„° ê·¸ë˜í”„ì— í‘œì‹œí•˜ê¸°
import matplotlib.pyplot as plt

plt.scatter(x_data, y_data)
plt.xlabel("YearsExperience")
plt.ylabel("Salary")
plt.show()
```

```python
# í›ˆë ¨ ë°ì´í„°, í‰ê°€ ë°ì´í„° ë¶„ë¦¬í•˜ê¸°
from sklearn.model_selection import train_test_split

x_train, x_val, y_train, y_val = train_test_split(x_data, y_data, test_size=0.4, random_state=2021)
print(x_train.shape, x_val.shape)
print(y_train.shape, y_val.shape)
```

```python
# ì„ í˜•íšŒê·€ ëª¨ë¸ í•™ìŠµ
from sklearn.linear_model import LinearRegression

model = LinearRegression(fit_intercept=True)
model.fit(x_train, y_train)
```

```python
print('yì ˆí¸ : ', model.intercept_)
print('ê°€ì¤‘ì¹˜ : ', model.coef_)
```

```python
# ì„ í˜•íšŒê·€ ê²°ê³¼ ì‹œê°í™”(í‰ê°€ ë°ì´í„°)
import matplotlib.pyplot as plt

y_pred = model.predict(x_val)

plt.scatter(x_val, y_val)
plt.xlabel("YearsExperience")
plt.ylabel("Salary")
plt.plot(x_val, y_pred, color="red")
plt.show()
```

```python
# ì„ í˜•íšŒê·€ ê²°ê³¼ ì‹œê°í™”(í‰ê°€ ë°ì´í„° + í›ˆë ¨ ë°ì´í„°)
import matplotlib.pyplot as plt

y = model.predict(x_data)

plt.scatter(x_data, y_data)
plt.xlabel("YearsExperience")
plt.ylabel("Salary")
plt.plot(x_data, y, color="red")
plt.show()
```

```python
from sklearn.metrics import mean_squared_error

print("í‰ê·  ì œê³± ì˜¤ì°¨: " + str(mean_squared_error(model.predict(x_val), y_val)))
```
