**1. ì •í™•ë„(Accuracy)**   

ì •í™•ë„ëŠ” ëª¨ë“  ë°ì´í„° ì¤‘ ëª¨ë¸ì´ ì˜¬ë°”ë¥´ê²Œ íŒë‹¨í•œ ë°ì´í„°ì˜ ë¹„ìœ¨ì´ì—ìš”.  
í‰ê°€ ì§€í‘œ ì¤‘ ê°€ì¥ ì§ê´€ì ì´ë©°, ê° í´ë˜ìŠ¤ì— ë°ì´í„°ê°€ **ê³ ë¥´ê²Œ ë¶„í¬ë˜ì–´ ìˆë‹¤ë©´** êµ‰ì¥íˆ ìœ ìš©í•œ ì§€í‘œì—ìš”.  
$$ Accuracy = \frac{TP + TN}{TP + TN + FP + FN}$$


**2. ì •ë°€ë„(Precision)**   

ì •ë°€ë„ëŠ” ëª¨ë¸ì´ ì°¸ìœ¼ë¡œ ì˜ˆì¸¡í•œ ëª¨ë“  ë°ì´í„° ì¤‘ì—ì„œ ì‹¤ì œë¡œ ì°¸ì¸ ë°ì´í„°ì˜ ë¹„ìœ¨ì´ì—ìš”.

$$ Precision = \frac{TP}{TP + FP}$$

ì¬í˜„ìœ¨ì€ ì‹¤ì œë¡œ ì–‘ì„±ì¸ ê²ƒ ì¤‘ì—ì„œ ëª¨ë¸ì´ ì–‘ì„±ìœ¼ë¡œ ì˜ˆì¸¡í•œ ë¹„ìœ¨ì´ì—ìš”.  
ì½”ë¡œë‚˜19 ì§„ë‹¨ê³¼ ê°™ì´, ì‹¤ì œë¡œëŠ” ì°¸ì¸ë° ê±°ì§“ìœ¼ë¡œ ì˜ˆì¸¡í•˜ëŠ” ê²½ìš°ë¥¼ ìµœì†Œí™”í•´ì•¼ í•  ë•Œ ì‚¬ìš©í•  ìˆ˜ ìˆì–´ìš”.

<h2 style='font-weight: bold'>3. ë¶„ë¥˜ëª¨ë¸ í‰ê°€ ì§€í‘œ ì‹¤ìŠµ</h2>

ì‹¤ìŠµì—ì„œëŠ” ìœ ë°©ì•” ì§„ë‹¨ ë°ì´í„°ë¥¼ ì˜ì‚¬ê²°ì •íŠ¸ë¦¬ ëª¨ë¸ë¡œ í•™ìŠµì‹œí‚¬ ê±°ì—ìš”.  
í•™ìŠµí•œ ëª¨ë¸ì„ í™œìš©í•´, í‰ê°€ ë°ì´í„°ì— ëŒ€í•´ í˜¼ë™ í–‰ë ¬ì„ êµ¬í•œ ë’¤ í‰ê°€ ì§€í‘œë¥¼ ê³„ì‚°í•´ë³´ì•„ìš”.

```python
# ìœ ë°©ì•” ë°ì´í„° ë¶ˆëŸ¬ì˜¤ê¸°
import numpy as np
import pandas as pd
from sklearn.datasets import load_breast_cancer # ìœ ë°©ì•” ìœ ë¬´ íŒë³„ ë°ì´í„°ë¥¼ ë¶ˆëŸ¬ì˜¤ëŠ” í•¨ìˆ˜
from sklearn.model_selection import train_test_split

X, Y = load_breast_cancer(return_X_y = True)
X = np.array(X)
Y = np.array(Y)

train_X, test_X, train_Y, test_Y = train_test_split(X, Y, test_size=0.2, random_state = 42)
```

```python
# ì˜ì‚¬ê²°ì •íŠ¸ë¦¬ í•™ìŠµ ë° ì˜ˆì¸¡
from sklearn.tree import DecisionTreeClassifier

DTmodel = DecisionTreeClassifier()
DTmodel.fit(train_X, train_Y)

y_pred_test = DTmodel.predict(test_X)
```

```python
# ì˜ˆì¸¡í•œ ê°’ê³¼ ì‹¤ì œ ê°’ì„ í† ëŒ€ë¡œ í˜¼ë™ í–‰ë ¬ êµ¬ì„±
from sklearn.metrics import confusion_matrix

cm_test = confusion_matrix(test_Y, y_pred_test)
print('í˜¼ë™ í–‰ë ¬(Confusion Matrix) : \n {}'.format(cm_test),'\n')
```

```python
# ì—¬ëŸ¬ ê°€ì§€ í‰ê°€ ì§€í‘œ ê³„ì‚°
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score

acc_test = accuracy_score(test_Y, y_pred_test)
print('ì •í™•ë„(Accuracy): %f' % (acc_test))

precision_test = precision_score(test_Y, y_pred_test)
print('ì •ë°€ë„(Precision): %f' % (precision_test))

recall_test = recall_score(test_Y, y_pred_test)
print('ì¬í˜„ìœ¨(Recall): %f' % (recall_test))
```
